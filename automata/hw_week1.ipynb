## *Theorem*

- $\log_2{3}\neq \frac{n}{m},\ \ n,m \in \mathbb{Z}^+$

## *Proof by Contradiction*

Assume the opposite of our theorem:
- $\log_2{3}= \frac{n}{m},\ \ n,m \in \mathbb{Z}^+$

### - Simplify Equation -


Using the principles of logarithms, simplify the equation: 
- $\log_2{3}= \frac{n}{m}$
- $2^\frac{n}{m} = 3$

       
Multiply both sides of the equation by the power of $m$ to remove the fraction in the exponent: 
- $2^n = 3^m$

### - Prime Factorization -

According to the Fundamental Theorem of Arithmetic, prime numbers have no divisors other than 1 and themselves.
  
The only prime factors of $2^n$ are $2$
- $2^n = 2 \times 2 \times \dots \times 2$
  
The only prime factors of $3^m$ are $3$
- $3^m = 3 \times 3 \times \dots \times 3$

### - Conclusion -

$2$ and $3$ have no common prime factors.  
So they have no common multiple other than 1.  
Sicne $n,m \in \mathbb{Z}^+$, a result of 1 cannot be reached with $2^0=3^0$.  


$\therefore \ $ The proof by contradiction is $false$:
- $\log_2{3}= \frac{n}{m},\ \ n,m \in \mathbb{Z}^+$

**Thus proving our original theorem:**
- $\log_2{3}\neq \frac{n}{m},\ \ n,m \in \mathbb{Z}^+$

## *Theorem*

- $\sum_{i=0}^{n} 2^i = 2^{n + 1} - 1,\ \ n \in \mathbb{N}$


## *Proof by Induction*

### - Base Case -

Verify $n=0$ is true.
  
*For the **right** hand side:*
- $\sum_{i=0}^{0} 2^i$
- $2^0$
- $1$
  
*For the **left** hand side:*
- $2^{0+1} - 1$
- $2^1 - 1$
- $1$



Since $1=1$ is $true$, the base case holds.

### - Inductive Hypothesis - 

Assume $\sum_{i=0}^{k} 2^i = 2^{k + 1} - 1$

where $n = k, \ \forall k \in \mathbb{N}$

The theorem holds for $n = k$. Next, we'll show it holds for $n = k + 1$:
- $\sum_{i=0}^{k + 1} 2^i = 2^{ \left( k + 1 \right) + 1 } - 1$

### - Inductive Step - 

The theorem holds for $n = k$. Next we'll show it holds for $n = k + 1$

- The summation of $i = 0$  to  $i = k$ can be expressed with $\sum_{i=0}^{k} 2^i$
  
- The final iteration of the summantion where $i = k + 1$ can be expressed as $2^{k+1}$
  
To express the summation where $n = k + 1$, we can write:
  
- $\sum_{i=0}^{k+1} 2^i = \left( \sum_{i=0}^{k} 2^i \right) + 2^{k+1}$

From the *Inductive Hypothesis* where $n = k$, we know:
  
- $\sum_{i=0}^{k} 2^i = 2^{k + 1} - 1$
  
In the equation shown above, we can substitute $\ \left( \sum_{i=0}^{k} 2^i \right)\ $ with $\ 2^{k + 1} - 1$, giving us:
- $\sum_{i=0}^{k+1} 2^i = \left( 2^{k + 1} - 1 \right) + 2^{k+1}$
  
Now we can simplify:
- $\left( 2^{k+1} - 1 \right) + 2^{k+1}$
- $2^{k+1} + 2^{k+1} - 1$
- $2^{k+2} - 1$

If we take the right side of the equation from the *Inductiive Hypothesis*:
- $\ \sum_{i=0}^{k + 1} 2^i = 2^{ \left( k + 1 \right) +1 } - 1$
- $ 2^{ \left( k + 1 \right) +1 } - 1$
  
And substitue it equation into the left side of this equation:
- $\sum_{i=0}^{k+1} 2^i = \left( 2^{k + 1} - 1 \right) + 2^{k+1}$

We get:
- $\left( 2^{k+1} - 1 \right) + 2^{k+1} = \left( 2^{k + 1} - 1 \right) + 2^{k+1}$ 

### - Conclusion -

The formula holds for $n=k$, then it also holds for $n=k+1$.
  
Since the formula held for the base case where $n=0$, by mathematical induction the theorem is $true$ $ \forall n \in \mathbb{N}$
